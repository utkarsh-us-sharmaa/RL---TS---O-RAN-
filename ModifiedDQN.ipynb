```python
import numpy as np
import pandas as pd
import random
import torch
import torch.nn as nn
import torch.optim as optim
import torch.autograd as autograd
import torch.nn.functional as F
from collections import deque
import matplotlib.pyplot as plt

# Enable CUDA if available
USE_CUDA = torch.cuda.is_available()
dtype = torch.cuda.FloatTensor if USE_CUDA else torch.FloatTensor

# Policy for preference-based reward
policy = {
    '1': {'p1': 'macro', 'p2': 'pico', 'p3': 'anchored'},
    '2': {'p1': 'macro', 'p2': 'anchored', 'p3': 'pico'},
    '3': {'p1': 'pico', 'p2': 'anchored', 'p3': 'macro'},
    '4': {'p1': 'anchored', 'p2': 'macro', 'p3': 'pico'}
}

class Environment:
    def __init__(self, ues, cells, policy):
        self.ues = ues.reset_index(drop=True)
        self.cells = cells.reset_index(drop=True)
        self.num_ues = len(ues)
        self.num_cells = len(cells)
        self.policy = policy  # Pass in the policy for preference-based reward calculation
        self.reset()

    def reset(self):
        self.current_step = 0
        self.ues['allocatedCell'] = 0
        self.cells['currCapacity'] = self.cells['maxCapacity-Kbps']
        return self.get_state()

    def step(self, action):
        cell_id = action  # action is cell_id
        ue_id = self.current_step  # current UE index

        ue = self.ues.iloc[ue_id]
        cell = self.cells.iloc[cell_id]
        reward = 0

        # Determine the cell type
        if cell['anchored'] == 1:
            cellType = "anchored"
        elif cell['pico'] == 1:
            cellType = "pico"
        else:
            cellType = "macro"

        # Get the preferences for the UE slice
        ue_slice = str(int(ue['slice']))
        preference = self.policy[ue_slice]

        # Check if the cell can accommodate the UE's throughput
        if cell['currCapacity'] >= ue['throughput(Kbps)']:
            self.cells.loc[cell_id, 'currCapacity'] -= ue['throughput(Kbps)']
            self.ues.loc[ue_id, 'allocatedCell'] = cell_id + 1  # allocate the cell

            # Reward based on how well the cell type matches the preference
            if cellType == preference['p1']:
                reward = 100  # Highest reward for first preference
            elif cellType == preference['p2']:
                reward = 50   # Medium reward for second preference
            elif cellType == preference['p3']:
                reward = 25   # Lowest reward for third preference

            # Bonus reward for efficient utilization
            utilization = 1 - (cell['currCapacity'] / cell['maxCapacity-Kbps'])
            if utilization < 0.8:
                reward += 5

        else:
            reward = -2  # Penalty for overloading a base station

        # Penalty for over-utilized cells
        utilization = 1 - (cell['currCapacity'] / cell['maxCapacity-Kbps'])
        if utilization > 0.9:
            reward -= 10

        self.current_step += 1
        done = self.current_step >= len(self.ues)
        next_state = self.get_state()
        return next_state, reward, done, {}

    def get_state(self):
        # Cell capacities (normalized)
        cell_capacities = self.cells['currCapacity'].values / self.cells['maxCapacity-Kbps']

        # Current UE info
        if self.current_step < self.num_ues:
            current_ue = self.ues.iloc[self.current_step]
            ue_throughput = current_ue['throughput(Kbps)'] / self.ues['throughput(Kbps)'].max()
            ue_slice = int(current_ue['slice']) - 1  # slices are 1-4
            ue_slice_encoded = np.zeros(4)
            ue_slice_encoded[ue_slice] = 1
        else:
            # If all UEs have been processed, return zeros
            ue_throughput = 0
            ue_slice_encoded = np.zeros(4)

        state = np.concatenate((cell_capacities, [ue_throughput], ue_slice_encoded))
        return state

    def get_unmanaged(self):
        return (self.ues['allocatedCell'] == 0).sum()

class DQN(nn.Module):
    def __init__(self, state_size, action_size):
        super(DQN, self).__init__()
        self.fc1 = nn.Linear(state_size, 128)
        self.fc2 = nn.Linear(128, 128)
        self.fc3 = nn.Linear(128, 64)
        self.fc4 = nn.Linear(64, action_size)

    def forward(self, x):
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = F.relu(self.fc3(x))
        return self.fc4(x)

class DQNAgent:
    def __init__(self, state_size, action_size):
        self.state_size = state_size
        self.action_size = action_size
        self.memory = deque(maxlen=50000)
        self.gamma = 0.95
        self.epsilon = 1.0
        self.epsilon_min = 0.01
        self.epsilon_decay = 0.995
        self.learning_rate = 0.00005
        self.model = DQN(state_size, action_size)
        self.target_model = DQN(state_size, action_size)
        self.update_target_model()
        self.optimizer = optim.Adam(self.model.parameters(), lr=self.learning_rate)

    def update_target_model(self):
        self.target_model.load_state_dict(self.model.state_dict())

    def remember(self, state, action, reward, next_state, done):
        self.memory.append((state, action, reward, next_state, done))

    def act(self, state):
        if np.random.rand() <= self.epsilon:
            return random.randrange(self.action_size)
        state = torch.FloatTensor(state).unsqueeze(0)
        act_values = self.model(state)
        return torch.argmax(act_values[0]).item()

    def replay(self, batch_size):
        if len(self.memory) < batch_size:
            return None
        minibatch = random.sample(self.memory, batch_size)
        states, actions, rewards, next_states, dones = zip(*minibatch)
        states = torch.FloatTensor(states)
        actions = torch.LongTensor(actions)
        rewards = torch.FloatTensor(rewards)
        next_states = torch.FloatTensor(next_states)
        dones = torch.FloatTensor(dones)

        Q_targets_next = self.target_model(next_states).detach().max(1)[0]
        Q_targets = rewards + (self.gamma * Q_targets_next * (1 - dones))
        Q_expected = self.model(states).gather(1, actions.unsqueeze(1)).squeeze(1)

        loss = F.mse_loss(Q_expected, Q_targets)
        self.optimizer.zero_grad()
        loss.backward()
        self.optimizer.step()

        return loss.item()

def train_dqn(ues, cells, policy, episodes=200, batch_size=64):
    env = Environment(ues, cells, policy)
    state_size = len(env.get_state())
    action_size = env.num_cells  # number of cells
    agent = DQNAgent(state_size, action_size)
    episodic_rewards = []
    unmanaged_ues_list = []
    losses = []

    for e in range(episodes):
        state = env.reset()
        total_reward = 0
        while True:
            action = agent.act(state)
            next_state, reward, done, info = env.step(action)
            agent.remember(state, action, reward, next_state, done)
            state = next_state
            total_reward += reward
            if done:
                break
            if len(agent.memory) > batch_size:
                loss = agent.replay(batch_size)
                if loss is not None:
                    losses.append(loss)
        # Decrease epsilon
        if agent.epsilon > agent.epsilon_min:
            agent.epsilon *= agent.epsilon_decay

        unmanaged_ues = env.get_unmanaged()
        unmanaged_ues_list.append(unmanaged_ues)
        episodic_rewards.append(total_reward)
        print(f'Episode {e+1}/{episodes}: Total reward = {total_reward}, Unmanaged UEs = {unmanaged_ues}, Epsilon = {agent.epsilon}')

    return agent, episodic_rewards, losses, unmanaged_ues_list

def test_agent(agent, ues, cells, policy, episodes=1):
    env = Environment(ues, cells, policy)
    unmanaged_ues_list = []
    total_rewards = []

    agent.epsilon = 0  # Disable exploration, test using learned policy

    for e in range(episodes):
        state = env.reset()
        total_reward = 0
        while True:
            action = agent.act(state)
            next_state, reward, done, info = env.step(action)
            state = next_state
            total_reward += reward
            if done:
                break

        unmanaged_ues = env.get_unmanaged()
        unmanaged_ues_list.append(unmanaged_ues)
        total_rewards.append(total_reward)
        print(f'Test Episode {e+1}/{episodes}: Total reward = {total_reward}, Unmanaged UEs = {unmanaged_ues}')

    return total_rewards, unmanaged_ues_list

def plot_results(episodic_rewards, losses, unmanaged_ues_list):
    plt.figure(figsize=(18, 5))

    plt.subplot(1, 3, 1)
    plt.plot(episodic_rewards)
    plt.xlabel('Episode')
    plt.ylabel('Total Reward')
    plt.title('Episodic Rewards')

    plt.subplot(1, 3, 2)
    plt.plot(losses)
    plt.xlabel('Training Step')
    plt.ylabel('Loss')
    plt.title('Training Losses')

    plt.subplot(1, 3, 3)
    plt.plot(unmanaged_ues_list)
    plt.xlabel('Episode')
    plt.ylabel('Unmanaged UEs')
    plt.title('Unmanaged UEs per Episode')

    plt.tight_layout()
    plt.show()

# Example usage
ues = pd.read_csv('./data/data.csv')  # Provide the correct path to your data file
cells = pd.read_csv('./data/cells.csv')  # Provide the correct path to your cells file

agent, rewards, losses, unmanaged_ues_list = train_dqn(ues, cells, policy, episodes=200, batch_size=64)
plot_results(rewards, losses, unmanaged_ues_list)

test_rewards, test_unmanaged_ues = test_agent(agent, ues, cells, policy, episodes=10)
```

```python
#THIS IS IT
import numpy as np
import pandas as pd
import random
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
from collections import deque
import matplotlib.pyplot as plt
import os

# Set random seeds for reproducibility
np.random.seed(42)
torch.manual_seed(42)
random.seed(42)

# Enable CUDA if available
USE_CUDA = torch.cuda.is_available()
device = torch.device("cuda" if USE_CUDA else "cpu")

# Policy for preference-based reward
policy = {
    '1': {'p1': 'macro', 'p2': 'pico', 'p3': 'anchored'},
    '2': {'p1': 'macro', 'p2': 'anchored', 'p3': 'pico'},
    '3': {'p1': 'pico', 'p2': 'anchored', 'p3': 'macro'},
    '4': {'p1': 'anchored', 'p2': 'macro', 'p3': 'pico'}
}

class Environment:
    def __init__(self, ues, cells, policy):
        self.ues_original = ues.copy()  # Keep a copy of the original UEs
        self.cells_original = cells.copy()  # Keep a copy of the original cells
        self.policy = policy  # Policy for preference-based reward calculation
        self.num_ues = len(ues)
        self.num_cells = len(cells)
        self.reset()

    def reset(self):
        self.current_step = 0
        # Optionally shuffle UEs to prevent overfitting to a specific order
        # self.ues = self.ues_original.sample(frac=1).reset_index(drop=True)
        self.ues = self.ues_original.copy().reset_index(drop=True)
        self.ues['allocatedCell'] = 0
        self.cells = self.cells_original.copy()
        self.cells['currCapacity'] = self.cells['maxCapacity-Kbps'].astype(float)
        return self.get_state()
    
    def step(self, action):
        cell_id = action  # action is cell_id
        ue_id = self.current_step  # current UE index

        ue = self.ues.iloc[ue_id]
        cell = self.cells.iloc[cell_id]
        reward = 0

        # Determine the cell type
        if cell['anchored'] == 1:
            cellType = "anchored"
        elif cell['pico'] == 1:
            cellType = "pico"
        else:
            cellType = "macro"

        # Get the preferences for the UE slice
        ue_slice = str(int(ue['slice']))
        preference = self.policy[ue_slice]

        # Use actual values for throughput and capacities
        ue_throughput = ue['throughput(Kbps)']
        cell_capacity = cell['currCapacity']
        if cell_capacity >= ue_throughput:
            self.cells.loc[cell_id, 'currCapacity'] -= ue_throughput
            self.ues.loc[ue_id, 'allocatedCell'] = cell_id + 1  # allocate the cell             

            if cellType == preference['p1']:
                reward = 0.1  # Reduced reward
            elif cellType == preference['p2']:
                reward = 0.05
            elif cellType == preference['p3']:
                reward = 0.01

            minIdx = list(self.cells['currCapacity']).index(min(self.cells['currCapacity']))
            maxCurrLoad = (self.cells['maxCapacity-Kbps'][minIdx]-self.cells['currCapacity'][minIdx])/self.cells['maxCapacity-Kbps'][minIdx]
            if maxCurrLoad > 0.95:
                reward -= 0.1  # Adjusted penalty
            else:
                reward += 0.02  # Adjusted bonus
        else:
            reward -= 0.1  # Adjusted penalty
            self.ues.loc[ue_id, 'allocatedCell'] = 0

        self.current_step += 1
        done = self.current_step >= len(self.ues)

        next_state = self.get_state()
        return next_state, reward, done, {}

    def get_state(self):
        # Cell capacities (normalized)
        cell_capacities = self.cells['currCapacity'].values / self.cells['maxCapacity-Kbps'].values

        # Current UE info
        if self.current_step < self.num_ues:
            current_ue = self.ues.iloc[self.current_step]
            ue_throughput = current_ue['throughput(Kbps)'] / self.ues['throughput(Kbps)'].max()
            ue_slice = int(current_ue['slice']) - 1  # slices are 1-4
            ue_slice_encoded = np.zeros(4)
            ue_slice_encoded[ue_slice] = 1
        else:
            # If all UEs have been processed, return zeros
            ue_throughput = 0
            ue_slice_encoded = np.zeros(4)

        state = np.concatenate((cell_capacities, [ue_throughput], ue_slice_encoded))
        return state

    def get_unmanaged(self):
        return (self.ues['allocatedCell'] == 0).sum()

class DuelingDQN(nn.Module):
    def __init__(self, state_size, action_size):
        super(DuelingDQN, self).__init__()
        self.fc1 = nn.Sequential(
            nn.Linear(state_size, 256),
            nn.ReLU()
        )
        self.fc2 = nn.Sequential(
            nn.Linear(256, 256),
            nn.ReLU()
        )

        # Value and Advantage streams
        self.value_stream = nn.Linear(256, 1)
        self.advantage_stream = nn.Linear(256, action_size)

    def forward(self, x):
        x = self.fc1(x)
        x = self.fc2(x)

        value = self.value_stream(x)
        advantage = self.advantage_stream(x)

        q_vals = value + (advantage - advantage.mean(dim=1, keepdim=True))
        return q_vals

class DQNAgent:
    def __init__(self, state_size, action_size):
        self.state_size = state_size
        self.action_size = action_size
        self.memory = deque(maxlen=100000)
        self.gamma = 0.9  # Reduced discount factor
        self.epsilon = 1.0  # Exploration rate
        self.epsilon_min = 0.01
        self.epsilon_decay = 0.995
        self.learning_rate = 1e-5  # Further reduced learning rate
        self.model = DuelingDQN(state_size, action_size).to(device)
        self.target_model = DuelingDQN(state_size, action_size).to(device)
        self.update_target_model()
        self.optimizer = optim.Adam(self.model.parameters(), lr=self.learning_rate)
        self.update_counter = 0  # For updating target model periodically
        self.q_values = []  # To track average Q-values

    def update_target_model(self):
        self.target_model.load_state_dict(self.model.state_dict())

    def remember(self, state, action, reward, next_state, done):
        # Store experience in memory
        self.memory.append((state, action, reward, next_state, done))

    def act(self, state):
        # Epsilon-greedy action selection
        if np.random.rand() <= self.epsilon:
            return random.randrange(self.action_size)
        state = torch.FloatTensor(state).unsqueeze(0).to(device)
        with torch.no_grad():
            act_values = self.model(state)
        return torch.argmax(act_values[0]).item()

    def replay(self, batch_size):
        if len(self.memory) < batch_size:
            return None
        minibatch = random.sample(self.memory, batch_size)
        states, actions, rewards, next_states, dones = zip(*minibatch)
        states = torch.FloatTensor(np.array(states)).to(device)
        actions = torch.LongTensor(actions).unsqueeze(1).to(device)
        rewards = torch.FloatTensor(rewards).unsqueeze(1).to(device)
        next_states = torch.FloatTensor(np.array(next_states)).to(device)
        dones = torch.FloatTensor(dones).unsqueeze(1).to(device)

        # Compute Q targets for current states
        Q_expected = self.model(states).gather(1, actions)

        # Double DQN
        with torch.no_grad():
            next_actions = self.model(next_states).argmax(1).unsqueeze(1)
            Q_targets_next = self.target_model(next_states).gather(1, next_actions)

        Q_targets = rewards + (self.gamma * Q_targets_next * (1 - dones))

        # Compute loss using Huber loss
        loss = F.smooth_l1_loss(Q_expected, Q_targets)

        # Optimize the model
        self.optimizer.zero_grad()
        loss.backward()
        # Reintroduce gradient clipping
        torch.nn.utils.clip_grad_norm_(self.model.parameters(), max_norm=1.0)
        self.optimizer.step()

        # Update target network more frequently
        self.update_counter += 1
        if self.update_counter % 500 == 0:
            self.update_target_model()

        # Track average Q-value
        average_Q_value = Q_expected.mean().item()
        self.q_values.append(average_Q_value)

        return loss.item()

def train_dqn(ues, cells, policy, episodes=200, batch_size=128):
    env = Environment(ues, cells, policy)
    state_size = len(env.get_state())
    action_size = env.num_cells  # number of cells
    agent = DQNAgent(state_size, action_size)
    episodic_rewards = []
    unmanaged_ues_list = []
    losses = []
    steps = 0

    # Check total capacity and total throughput demand
    total_cell_capacity = cells['maxCapacity-Kbps'].sum()
    total_ue_throughput = ues['throughput(Kbps)'].sum()
    print(f"Total Cell Capacity: {total_cell_capacity}")
    print(f"Total UE Throughput Demand: {total_ue_throughput}")

    best_unmanaged_ues = float('inf')

    for e in range(episodes):
        state = env.reset()
        total_reward = 0
        while True:
            action = agent.act(state)
            next_state, reward, done, _ = env.step(action)
            agent.remember(state, action, reward, next_state, done)
            state = next_state
            total_reward += reward
            steps += 1

            if len(agent.memory) > batch_size:
                loss = agent.replay(batch_size)
                if loss is not None:
                    losses.append(loss)

            if done:
                break

        # Decrease epsilon
        if agent.epsilon > agent.epsilon_min:
            agent.epsilon *= agent.epsilon_decay

        unmanaged_ues = env.get_unmanaged()
        unmanaged_ues_list.append(unmanaged_ues)
        episodic_rewards.append(total_reward)

        # Model checkpointing
        if unmanaged_ues < best_unmanaged_ues:
            best_unmanaged_ues = unmanaged_ues
            torch.save(agent.model.state_dict(), 'best_model.pth')
            print(f'New best model saved with {unmanaged_ues} unmanaged UEs.')

        print(f'Episode {e+1}/{episodes}: Total reward = {total_reward:.4f}, Unmanaged UEs = {unmanaged_ues}, Epsilon = {agent.epsilon:.4f}')

    # Save the final model
    torch.save(agent.model.state_dict(), 'final_model.pth')
    print('Final model saved.')

    return agent, episodic_rewards, losses, unmanaged_ues_list

def test_agent(agent, ues, cells, policy, episodes=5, model_path='best_model.pth'):
    env = Environment(ues, cells, policy)
    unmanaged_ues_list = []
    total_rewards = []

    # Load the model
    if os.path.exists(model_path):
        agent.model.load_state_dict(torch.load(model_path, map_location=device))
        print(f"Loaded model from {model_path} for testing.")

    agent.epsilon = 0  # Disable exploration, test using learned policy

    for e in range(episodes):
        state = env.reset()
        total_reward = 0
        while True:
            action = agent.act(state)
            next_state, reward, done, _ = env.step(action)
            state = next_state
            total_reward += reward
            if done:
                break

        unmanaged_ues = env.get_unmanaged()
        unmanaged_ues_list.append(unmanaged_ues)
        total_rewards.append(total_reward)
        print(f'Test Episode {e+1}/{episodes}: Total reward = {total_reward:.4f}, Unmanaged UEs = {unmanaged_ues}')

    avg_unmanaged_ues = np.mean(unmanaged_ues_list)
    print(f'Average Unmanaged UEs over {episodes} test episodes: {avg_unmanaged_ues}')
    return total_rewards, unmanaged_ues_list

def plot_results(episodic_rewards, losses, unmanaged_ues_list, q_values):
    plt.figure(figsize=(24, 5))

    plt.subplot(1, 4, 1)
    plt.plot(episodic_rewards)
    plt.xlabel('Episode')
    plt.ylabel('Total Reward')
    plt.title('Episodic Rewards')

    plt.subplot(1, 4, 2)
    # Plot moving average of losses for smoother curve
    moving_avg = pd.Series(losses).rolling(window=100).mean()
    plt.plot(moving_avg)
    plt.xlabel('Training Step')
    plt.ylabel('Loss')
    plt.title('Training Losses (Moving Average)')

    plt.subplot(1, 4, 3)
    plt.plot(unmanaged_ues_list)
    plt.xlabel('Episode')
    plt.ylabel('Unmanaged UEs')
    plt.title('Unmanaged UEs per Episode')

    plt.subplot(1, 4, 4)
    plt.plot(q_values)
    plt.xlabel('Training Step')
    plt.ylabel('Average Q-Value')
    plt.title('Average Q-Values During Training')

    plt.tight_layout()
    plt.show()

# Load data
ues = pd.read_csv('./data/data.csv')  # Replace with your actual data file path
cells = pd.read_csv('./data/cells.csv')  # Replace with your actual cells file path

# Ensure 'throughput(Kbps)' and 'maxCapacity-Kbps' are of type float
ues['throughput(Kbps)'] = ues['throughput(Kbps)'].astype(float)
cells['maxCapacity-Kbps'] = cells['maxCapacity-Kbps'].astype(float)

# Run training
agent, rewards, losses, unmanaged_ues_list = train_dqn(ues, cells, policy, episodes=200, batch_size=128)

# Plot results
plot_results(rewards, losses, unmanaged_ues_list, agent.q_values)

# Evaluate agent with epsilon=0 after training
print('Evaluating agent with epsilon=0 after training:')
agent.epsilon = 0
test_rewards, test_unmanaged_ues = test_agent(agent, ues, cells, policy, episodes=5, model_path='final_model.pth')

# Run testing with best model
print('Testing with best model:')
test_rewards_best, test_unmanaged_ues_best = test_agent(agent, ues, cells, policy, episodes=5, model_path='best_model.pth')
```

```python
import numpy as np
import pandas as pd
import random
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
from collections import deque
import matplotlib.pyplot as plt
import os

# Set random seeds for reproducibility
np.random.seed(42)
torch.manual_seed(42)
random.seed(42)

# Enable CUDA if available
USE_CUDA = torch.cuda.is_available()
device = torch.device("cuda" if USE_CUDA else "cpu")

# Policy for preference-based reward
policy = {
    '1': {'p1': 'macro', 'p2': 'pico', 'p3': 'anchored'},
    '2': {'p1': 'macro', 'p2': 'anchored', 'p3': 'pico'},
    '3': {'p1': 'pico', 'p2': 'anchored', 'p3': 'macro'},
    '4': {'p1': 'anchored', 'p2': 'macro', 'p3': 'pico'}
}

class Environment:
    def __init__(self, ues, cells, policy):
        self.ues_original = ues.copy()  # Keep a copy of the original UEs
        self.cells_original = cells.copy()  # Keep a copy of the original cells
        self.policy = policy  # Policy for preference-based reward calculation
        self.num_ues = len(ues)
        self.num_cells = len(cells)
        self.reset()

    def reset(self):
        self.current_step = 0
        # Shuffle UEs to prevent overfitting to a specific order
        self.ues = self.ues_original.sample(frac=1).reset_index(drop=True)
        self.ues['allocatedCell'] = 0
        self.cells = self.cells_original.copy()
        self.cells['currCapacity'] = self.cells['maxCapacity-Kbps'].astype(float)
        return self.get_state()
    
    def step(self, action):
        cell_id = action  # action is cell_id
        ue_id = self.current_step  # current UE index

        ue = self.ues.iloc[ue_id]
        cell = self.cells.iloc[cell_id]
        reward = 0

        # Determine the cell type
        if cell['anchored'] == 1:
            cellType = "anchored"
        elif cell['pico'] == 1:
            cellType = "pico"
        else:
            cellType = "macro"

        # Get the preferences for the UE slice
        ue_slice = str(int(ue['slice']))
        preference = self.policy[ue_slice]

        # Use actual values for throughput and capacities
        ue_throughput = ue['throughput(Kbps)']
        cell_capacity = cell['currCapacity']
        if cell_capacity >= ue_throughput:
            self.cells.loc[cell_id, 'currCapacity'] -= ue_throughput
            self.ues.loc[ue_id, 'allocatedCell'] = cell_id + 1  # allocate the cell             

            if cellType == preference['p1']:
                reward = 1.0  # Highest reward
            elif cellType == preference['p2']:
                reward = 0.5  # Medium reward
            elif cellType == preference['p3']:
                reward = 0.1  # Lowest reward

            minIdx = list(self.cells['currCapacity']).index(min(self.cells['currCapacity']))
            maxCurrLoad = (self.cells['maxCapacity-Kbps'][minIdx]-self.cells['currCapacity'][minIdx])/self.cells['maxCapacity-Kbps'][minIdx]
            if maxCurrLoad > 0.95:
                reward -= 1.0  # Heavily penalize overloading
            else:
                reward += 0.2  # Small bonus for keeping load balanced
        else:
            reward -= 1.0  # Penalize for attempting to allocate to over-capacity cell
            self.ues.loc[ue_id, 'allocatedCell'] = 0

        self.current_step += 1
        done = self.current_step >= len(self.ues)

        next_state = self.get_state()
        return next_state, reward, done, {}

    def get_state(self):
        # Cell capacities (normalized)
        cell_capacities = self.cells['currCapacity'].values / self.cells['maxCapacity-Kbps'].values

        # Current UE info
        if self.current_step < self.num_ues:
            current_ue = self.ues.iloc[self.current_step]
            ue_throughput = current_ue['throughput(Kbps)'] / self.ues['throughput(Kbps)'].max()
            ue_slice = int(current_ue['slice']) - 1  # slices are 1-4
            ue_slice_encoded = np.zeros(4)
            ue_slice_encoded[ue_slice] = 1
        else:
            # If all UEs have been processed, return zeros
            ue_throughput = 0
            ue_slice_encoded = np.zeros(4)

        state = np.concatenate((cell_capacities, [ue_throughput], ue_slice_encoded))
        return state

    def get_unmanaged(self):
        return (self.ues['allocatedCell'] == 0).sum()

class DuelingDQN(nn.Module):
    def __init__(self, state_size, action_size):
        super(DuelingDQN, self).__init__()
        self.fc1 = nn.Sequential(
            nn.Linear(state_size, 256),
            nn.ReLU(),
            nn.Dropout(0.05)
        )
        self.fc2 = nn.Sequential(
            nn.Linear(256, 256),
            nn.ReLU(),
            nn.Dropout(0.05)
        )

        # Value and Advantage streams
        self.value_stream = nn.Linear(256, 1)
        self.advantage_stream = nn.Linear(256, action_size)

    def forward(self, x):
        x = self.fc1(x)
        x = self.fc2(x)

        value = self.value_stream(x)
        advantage = self.advantage_stream(x)

        q_vals = value + (advantage - advantage.mean(dim=1, keepdim=True))
        return q_vals

class DQNAgent:
    def __init__(self, state_size, action_size):
        self.state_size = state_size
        self.action_size = action_size
        self.memory = deque(maxlen=100000)
        self.gamma = 0.99  # Discount factor
        self.epsilon = 1.0  # Exploration rate
        self.epsilon_min = 0.01
        self.epsilon_decay = 0.999
        self.learning_rate = 1e-3
        self.model = DuelingDQN(state_size, action_size).to(device)
        self.target_model = DuelingDQN(state_size, action_size).to(device)
        self.update_target_model()
        self.optimizer = optim.AdamW(self.model.parameters(), lr=self.learning_rate, weight_decay=1e-4)
        self.update_counter = 0  # For updating target model periodically

    def update_target_model(self):
        self.target_model.load_state_dict(self.model.state_dict())

    def remember(self, state, action, reward, next_state, done):
        # Store experience in memory
        self.memory.append((state, action, reward, next_state, done))

    def act(self, state):
        # Epsilon-greedy action selection
        if np.random.rand() <= self.epsilon:
            return random.randrange(self.action_size)
        state = torch.FloatTensor(state).unsqueeze(0).to(device)
        with torch.no_grad():
            act_values = self.model(state)
        return torch.argmax(act_values[0]).item()

    def replay(self, batch_size):
        if len(self.memory) < batch_size:
            return None
        minibatch = random.sample(self.memory, batch_size)
        states, actions, rewards, next_states, dones = zip(*minibatch)
        states = torch.FloatTensor(np.array(states)).to(device)
        actions = torch.LongTensor(actions).unsqueeze(1).to(device)
        rewards = torch.FloatTensor(rewards).unsqueeze(1).to(device)
        next_states = torch.FloatTensor(np.array(next_states)).to(device)
        dones = torch.FloatTensor(dones).unsqueeze(1).to(device)

        # Compute Q targets for current states
        Q_expected = self.model(states).gather(1, actions)

        # Double DQN
        with torch.no_grad():
            next_actions = self.model(next_states).argmax(1).unsqueeze(1)
            Q_targets_next = self.target_model(next_states).gather(1, next_actions)

        Q_targets = rewards + (self.gamma * Q_targets_next * (1 - dones))

        # Compute loss
        loss = F.mse_loss(Q_expected, Q_targets)

        # Optimize the model
        self.optimizer.zero_grad()
        loss.backward()
        torch.nn.utils.clip_grad_norm_(self.model.parameters(), max_norm=1.0)  # Gradient clipping
        self.optimizer.step()

        # Update target network periodically
        self.update_counter += 1
        if self.update_counter % 1000 == 0:
            self.update_target_model()

        return loss.item()

def train_dqn(ues, cells, policy, episodes=1000, batch_size=64):
    env = Environment(ues, cells, policy)
    state_size = len(env.get_state())
    action_size = env.num_cells  # number of cells
    agent = DQNAgent(state_size, action_size)
    episodic_rewards = []
    unmanaged_ues_list = []
    losses = []
    steps = 0

    # Check total capacity and total throughput demand
    total_cell_capacity = cells['maxCapacity-Kbps'].sum()
    total_ue_throughput = ues['throughput(Kbps)'].sum()
    print(f"Total Cell Capacity: {total_cell_capacity}")
    print(f"Total UE Throughput Demand: {total_ue_throughput}")

    best_unmanaged_ues = float('inf')

    for e in range(episodes):
        state = env.reset()
        total_reward = 0
        while True:
            action = agent.act(state)
            next_state, reward, done, _ = env.step(action)
            agent.remember(state, action, reward, next_state, done)
            state = next_state
            total_reward += reward
            steps += 1

            if len(agent.memory) > batch_size:
                loss = agent.replay(batch_size)
                if loss is not None:
                    losses.append(loss)

            if done:
                break

        # Decrease epsilon
        if agent.epsilon > agent.epsilon_min:
            agent.epsilon *= agent.epsilon_decay

        unmanaged_ues = env.get_unmanaged()
        unmanaged_ues_list.append(unmanaged_ues)
        episodic_rewards.append(total_reward)

        # Model checkpointing
        if unmanaged_ues < best_unmanaged_ues:
            best_unmanaged_ues = unmanaged_ues
            torch.save(agent.model.state_dict(), 'best_model.pth')
            print(f'New best model saved with {unmanaged_ues} unmanaged UEs.')

        print(f'Episode {e+1}/{episodes}: Total reward = {total_reward}, Unmanaged UEs = {unmanaged_ues}, Epsilon = {agent.epsilon:.4f}')

    # Save the final model
    torch.save(agent.model.state_dict(), 'final_model.pth')
    print('Final model saved.')

    return agent, episodic_rewards, losses, unmanaged_ues_list

def test_agent(agent, ues, cells, policy, episodes=10, model_path='best_model.pth'):
    env = Environment(ues, cells, policy)
    unmanaged_ues_list = []
    total_rewards = []

    # Load the model
    if os.path.exists(model_path):
        agent.model.load_state_dict(torch.load(model_path))
        print(f"Loaded model from {model_path} for testing.")

    agent.epsilon = 0  # Disable exploration, test using learned policy

    for e in range(episodes):
        state = env.reset()
        total_reward = 0
        while True:
            action = agent.act(state)
            next_state, reward, done, _ = env.step(action)
            state = next_state
            total_reward += reward
            if done:
                break

        unmanaged_ues = env.get_unmanaged()
        unmanaged_ues_list.append(unmanaged_ues)
        total_rewards.append(total_reward)
        print(f'Test Episode {e+1}/{episodes}: Total reward = {total_reward}, Unmanaged UEs = {unmanaged_ues}')

    avg_unmanaged_ues = np.mean(unmanaged_ues_list)
    print(f'Average Unmanaged UEs over {episodes} test episodes: {avg_unmanaged_ues}')
    return total_rewards, unmanaged_ues_list

def plot_results(episodic_rewards, losses, unmanaged_ues_list):
    plt.figure(figsize=(18, 5))

    plt.subplot(1, 3, 1)
    plt.plot(episodic_rewards)
    plt.xlabel('Episode')
    plt.ylabel('Total Reward')
    plt.title('Episodic Rewards')

    plt.subplot(1, 3, 2)
    plt.plot(losses)
    plt.xlabel('Training Step')
    plt.ylabel('Loss')
    plt.title('Training Losses')

    plt.subplot(1, 3, 3)
    plt.plot(unmanaged_ues_list)
    plt.xlabel('Episode')
    plt.ylabel('Unmanaged UEs')
    plt.title('Unmanaged UEs per Episode')

    plt.tight_layout()
    plt.show()

# Load data
ues = pd.read_csv('./data/data.csv')  # Replace with your actual data file path
cells = pd.read_csv('./data/cells.csv')  # Replace with your actual cells file path

# Ensure 'throughput(Kbps)' and 'maxCapacity-Kbps' are of type float
ues['throughput(Kbps)'] = ues['throughput(Kbps)'].astype(float)
cells['maxCapacity-Kbps'] = cells['maxCapacity-Kbps'].astype(float)

# Run training
agent, rewards, losses, unmanaged_ues_list = train_dqn(ues, cells, policy, episodes=200, batch_size=64)
plot_results(rewards, losses, unmanaged_ues_list)

# Evaluate agent with epsilon=0 after training
print('Evaluating agent with epsilon=0 after training:')
agent.epsilon = 0
test_rewards, test_unmanaged_ues = test_agent(agent, ues, cells, policy, episodes=5, model_path='final_model.pth')

# Run testing with best model
print('Testing with best model:')
test_rewards_best, test_unmanaged_ues_best = test_agent(agent, ues, cells, policy, episodes=5, model_path='best_model.pth')
```
